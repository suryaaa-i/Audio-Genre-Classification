```python
import requests
import json
import base64

# Function to get the access token
def get_access_token(client_id, client_secret):
    url = "https://accounts.spotify.com/api/token"
    headers = {
        "Content-Type": "application/x-www-form-urlencoded",
        "Authorization": f"Basic {base64.b64encode(f'{client_id}:{client_secret}'.encode()).decode()}"
    }
    data = {
        "grant_type": "client_credentials"
    }
    
    response = requests.post(url, headers=headers, data=data)
    token_info = response.json()
    
    # Check if 'access_token' is in the response
    if 'access_token' in token_info:
        return token_info['access_token']
    else:
        raise Exception("Could not retrieve access token: " + str(token_info))

# Function to retrieve tracks and their features
def get_tracks_data(genre, access_token):
    url = f"https://api.spotify.com/v1/search?q={genre}&type=track&market=US&limit=10"
    headers = {
        "Authorization": f"Bearer {access_token}"
    }
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        return response.json()
    else:
        raise Exception("Failed to retrieve tracks: " + response.text)

# Main execution flow
if __name__ == "__main__":
    CLIENT_ID = '78043ababf53490f806df29715082e32'  # Replace with your Client ID
    CLIENT_SECRET = '140c94a5c40a4e309ebad0e046062ceb'  # Replace with your Client Secret

    # Get access token
    access_token = get_access_token(CLIENT_ID, CLIENT_SECRET)
    
    # List of genres to retrieve data for
    genres = ['pop', 'rap', 'edm', 'r&b', 'country', 'rock']
    
    all_tracks = {}

    # Retrieve tracks for each genre
    for genre in genres:
        tracks_data = get_tracks_data(genre, access_token)
        all_tracks[genre] = tracks_data
    
    # Print the retrieved data for each genre
    for genre, data in all_tracks.items():
        print(f"Genre: {genre}")
        print(json.dumps(data, indent=2))
```
